```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
font = {'size' : 14}
import matplotlib
matplotlib.rc('font', **font)
import seaborn as sns
```

```python
df_internet_baf = pd.read_csv("datasets\Internet_BAF (1).csv")
df_internet_baf["Trimestre"] = df_internet_baf["Trimestre"].astype("category")
df_internet_baf["AÃ±o"] = df_internet_baf["AÃ±o"].astype("category")
df_internet_baf["Periodo"] = df_internet_baf["Periodo"].astype("category")
df_internet_baf["Banda ancha fija"] = df_internet_baf["Banda ancha fija"].str.replace(".", "").astype("int64")
df_internet_baf["Total"] = df_internet_baf["Total"].str.replace(".", "").astype("int64")
df_internet_baf["Dial up"] = df_internet_baf["Dial up"].apply(lambda x: int(x * 1000) if pd.notnull(x) else x)

### Agregar columnas al dataset

# Calcular el porcentaje de cada categoria en relaciÃ³n del total.
df_internet_baf["Dial_porcent"] = round((df_internet_baf["Dial up"] / df_internet_baf["Total"]) * 100, 2)
df_internet_baf["Banda_porcent"] = round((df_internet_baf["Banda ancha fija"] / df_internet_baf["Total"]) * 100, 2)

df_internet_baf = df_internet_baf.sort_index(ascending=False)

# Calcular el porcentaje de crecimiento Anual y Trimestral.
df_internet_baf['Porcentaje_Crecimiento_Trimestral'] = round(df_internet_baf['Total'].pct_change() * 100,2)

df_internet_baf['Porcentaje_Crecimiento_Anual'] = round((df_internet_baf['Total'] - df_internet_baf['Total'].shift(4)) / df_internet_baf['Total'].shift(4) * 100, 2)

# Guardar el archivo
df_internet_baf.to_csv("Transformados/internet_xcategoria_general.csv",index=True)
df_internet_baf
```

```python
df_acceso_per_vel = pd.read_csv("Internet_Accesos-por-velocidad (2).csv")
df_acceso_per_vel["AÃ±o"] = df_acceso_per_vel["AÃ±o"].astype("category")
df_acceso_per_vel["Trimestre"] = df_acceso_per_vel["Trimestre"].astype("category")
df_acceso_per_vel["Provincia"] = df_acceso_per_vel["Provincia"].astype("category")
df_acceso_per_vel["HASTA 512 kbps"] = df_acceso_per_vel["HASTA 512 kbps"].apply(lambda x: int(x * 1000))
df_acceso_per_vel["+ 512 Kbps - 1 Mbps"] = df_acceso_per_vel["+ 512 Kbps - 1 Mbps"].apply(lambda x: int(x * 1000))
df_acceso_per_vel["+ 1 Mbps - 6 Mbps"] = df_acceso_per_vel["+ 1 Mbps - 6 Mbps"].str.replace(".", "").astype("int64")
df_acceso_per_vel["+ 6 Mbps - 10 Mbps"] = df_acceso_per_vel["+ 6 Mbps - 10 Mbps"].apply(lambda x: int(x * 1000))
df_acceso_per_vel["+ 10 Mbps - 20 Mbps"] = df_acceso_per_vel["+ 10 Mbps - 20 Mbps"].apply(lambda x: int(x * 1000))
df_acceso_per_vel["+ 20 Mbps - 30 Mbps"] = df_acceso_per_vel["+ 20 Mbps - 30 Mbps"].apply(lambda x: int(x * 1000))
#df_acceso_per_vel["+ 512 Kbps - 1 Mbps"] = df_acceso_per_vel["+ 512 Kbps - 1 Mbps"].apply(lambda x: int(x * 1000))
df_acceso_per_vel["OTROS"] = df_acceso_per_vel["OTROS"].fillna(0)
df_acceso_per_vel["OTROS"] = df_acceso_per_vel["OTROS"].apply(lambda x: int(x * 1000 if x > 0 else 0))
df_acceso_per_vel["+ 30 Mbps"] = df_acceso_per_vel["+ 30 Mbps"].str.replace(".", "").astype("int64")
df_acceso_per_vel["Total"] = df_acceso_per_vel["Total"].str.replace(".", "").astype("int64")
df_acceso_per_vel["Total"]

df_acceso_per_vel.to_csv(r"Transformados\acceso_xvel_xperiodo_xprov.csv")
df_acceso_per_vel
```
